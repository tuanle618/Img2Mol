```python
import torch
from img2mol.inference import *
```

```python
from IPython.display import display
```

```python
from PIL import Image
```

```python
import os
```

```python
os.listdir("model/")
```

```python
device = "cuda:0" if torch.cuda.is_available() else "cpu"
img2mol = Img2MolInference(model_ckpt="model/model.ckpt",
                           device=device)
cddd_server = CDDDRequest()
```

```python
os.listdir("examples/")
```

```python
res = img2mol(filepath="examples/digital_example1.png", cddd_server=cddd_server)
```

```python
res
```

```python
input_img = Image.open(res["filepath"], "r")
```

```python
display(input_img)
```

# show prediction

```python
print(res["smiles"])
print()
res["mol"]
```
